```python
import numpy as np
import matplotlib as plt
```

```python
def friedmann_equation(t,a , h0,Omega_r,Omega_m,Omega_L):
    return h0 * a *(Omega_r * a**(-4) + Omega_m * a**(-3) + Omega_L)**0.5
```

```python
from scipy.integrate import solve_ivp
h0 = 2.19e-18 #s^-1
Omega_r = 9e-5;Omega_m = 0.315; Omega_L = 0.685 #Constants

#Time span (in s)
t_start = 0;
t_end = 4.35e17 # 13.9 Gyr in seconds

a0 = 1e-6

def friedmann(t,a):
    return h0 * a * np.sqrt(Omega_r * a**(-4)+Omega_m * a**(-3)+Omega_L)
sol = solve_ivp(friedmann, [t_start, t_end],[a0], t_eval = np.linspace(t_start,t_end,1000))
print(sol)
```

```python
import matplotlib.pyplot as plt

plt.plot(sol.t / 3.15e16, sol.y[0])
plt.xlabel("Time (Gyr)")
plt.ylabel("Scale factor a(t)")
plt.title("Evolution of the scale factor")
plt.grid(True)
plt.show()
```

```python
# The aove code showing how the universe expands from a=10^-6 to a=1 over 13.8 billion years
```

```python
a_vals = sol.y[0]
rho_r = Omega_r * a_vals**(-4)
rho_m = Omega_m * a_vals**(-3)
rho_L = Omega_L * np.ones_like(a_vals)
rho_total = rho_r + rho_m + rho_L
rho_r /= rho_total
rho_m /= rho_total
rho_L /= rho_total
plt.plot(sol.t / (3.15e16), rho_r, label='Radiation')
plt.plot(sol.t / (3.15e16), rho_m, label='Matter')
plt.plot(sol.t / (3.15e16), rho_L, label='Dark Energy')
plt.xlabel("Time (Gyr)")
plt.ylabel("Normalized Energy Density")
plt.title("Energy Density Fractions Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
