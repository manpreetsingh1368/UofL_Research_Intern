```python
from scipy.integrate import solve_ivp
h0 = 2.19e-18 #s^-1
Omega_r = 9e-5;Omega_m = 0.315; Omega_L = 0.685 #Constants

#Time span (in s)
t_start = 0;
t_end = 4.35e17 # 13.9 Gyr in seconds

a0 = 1e-6

def friedmann(t,a):
    return h0 * a * np.sqrt(Omega_r * a**(-4)+Omega_m * a**(-3)+Omega_L)
sol = solve_ivp(friedmann, [t_start, t_end],[a0], t_eval = np.linspace(t_start,t_end,1000))
print(sol)
```

```python
import matplotlib.pyplot as plt

plt.plot(sol.t / 3.15e16, sol.y[0])
plt.xlabel("Time (Gyr)")
plt.ylabel("Scale factor a(t)")
plt.title("Evolution of the scale factor")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# Constants
H0 = 67.4 * 1000 / (3.086e22)  # Hubble constant in 1/s
Omega_r = 9e-5
Omega_m = 0.315
Omega_L = 0.685

# Friedmann equation: da/dt = H0 * a * sqrt(...)
def friedmann(t, a):
    return H0 * a * np.sqrt(Omega_r * a**(-4) + Omega_m * a**(-3) + Omega_L)

# Time span and initial condition
t_span = (0, 5e17)  # seconds (~16 billion years)
a0 = 1e-8           # very small scale factor at early universe

# Solve the ODE
sol = solve_ivp(friedmann, t_span, [a0], t_eval=np.linspace(t_span[0], t_span[1], 1000))

# Extract solution
t = sol.t / (3.15e16)  # convert seconds to Gyr
a = sol.y[0]

# Energy densities
rho_r = Omega_r * a**(-4)
rho_m = Omega_m * a**(-3)
rho_L = Omega_L * np.ones_like(a)

# Normalize densities
rho_total = rho_r + rho_m + rho_L
rho_r /= rho_total
rho_m /= rho_total
rho_L /= rho_total

# Plot scale factor vs time
plt.figure(figsize=(10, 5))
plt.plot(t, a)
plt.xlabel('Time (Gyr)')
plt.ylabel('Scale factor a(t)')
plt.title('Evolution of Scale Factor in ΛCDM Universe')
plt.grid(True)
plt.show()

# Plot fractional energy densities
plt.figure(figsize=(10, 5))
plt.plot(a, rho_r, label='Radiation')
plt.plot(a, rho_m, label='Matter')
plt.plot(a, rho_L, label='Dark Energy')
plt.xscale('log')
plt.xlabel('Scale factor a')
plt.ylabel('Fractional Energy Density')
plt.title('Energy Densities vs Scale Factor')
plt.legend()
plt.grid(True)
plt.show()
```
